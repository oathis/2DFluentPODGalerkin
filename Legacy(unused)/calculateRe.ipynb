```python
#@code by oathis


trainRe=[200,300,400,500,600,700,800,900]
testRe=[100,150,200,250,300,350,400,450,500,550,600,650,700,750,800,850,900,950,1000]

tt = [trainRe,testRe]

rho = 998.2 # Density: kg/m^2
  
L = 0.01 # m
u_lid=0.1 #m/s



mu = 0.001003 # Viscosity : 0.001003 initiallizing

final_mu_list = []

for re_list in tt:
    mu_list = []  # 각 하위 리스트에 대한 mu_list 생성
    for Re in re_list:
        # Re = (rho * u_lid * L) / mu  =>  mu = (rho * u_lid * L) / Re
        mu = (rho * u_lid * L) / Re
        mu_list.append(mu)
    final_mu_list.append(mu_list)

print(final_mu_list[0])
print(final_mu_list[1])
```
